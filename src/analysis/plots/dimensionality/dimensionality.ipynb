```python
import os
_FINAL_DFS_DIR = os.path.join('..', '..', 'final_dfs')

import sys
sys.path.append('..')
import plotconfig
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import ticker
from matplotlib.ticker import FuncFormatter
from scipy import stats
import pandas as pd
import numpy as np
```

```python
dbf1 = os.path.join(_FINAL_DFS_DIR, 'results_dimensionality_eeg.parquet')
df1 = pd.read_parquet(dbf1, engine='pyarrow')

mask1 = ((df1['test_name'] == "random"))
df = df1[mask1]
df['max_score'] = df['scores'].apply(max)

plt.figure(figsize=(8, 6))

ax = sns.lineplot(data=df, 
             x='n_component', 
             y='pearsonr_statistic', 
             hue='method_name', 
             errorbar=('se', 1))

ax.axhline(y=plotconfig.IR_RANDOM_PEARSON_STATISTIC, color='r', linestyle='--', alpha=0.5, label='Theoretical Random')

ax.set_xlim([0, 50])
ax.invert_yaxis()

plt.title("PCA EEG - N_components vs Pearson statistics")

plotconfig.save_fig('pca_eeg')
```
